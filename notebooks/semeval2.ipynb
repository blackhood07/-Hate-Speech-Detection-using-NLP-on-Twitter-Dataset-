```python
stop_words = stopwords.words('english')
def remove_stopwords(string):
    word_list = [word.lower() for word in string.split()]
    stopwords_list = list(stopwords.words("english"))
    for word in word_list:
        if word in stopwords_list:
            word_list.remove(word)
    return ' '.join(word_list)
```

```python
msg['Text Transcription_length'] = msg['Text Transcription'].apply(len)
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'[a-zA-Z._]+\.[a-zA-Z]+',' ',str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub('\\n',' ',str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'\W',' ',str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'https\s+|www.\s+',r'', str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'http\s+|www.\s+',r'', str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'\s+[a-zA-Z]\s+',' ',str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'\^[a-zA-Z]\s+',' ',str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'\s+',' ',str(x)))
msg['Text Transcription'] = msg['Text Transcription'].str.lower()

msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\’", "\'", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"won\'t", "will not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"can\'t", "can not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"don\'t", "do not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"dont", "do not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"n\’t", " not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"n\'t", " not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'re", " are", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'s", " is", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\’d", " would", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\d", " would", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'ll", " will", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'t", " not", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'ve", " have", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'m", " am", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\n", "", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\r", "", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"[0-9]", "digit", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\'", "", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r"\"", "", str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'[?|!|\'|"|#]',r'', str(x)))
msg['Text Transcription'] = msg['Text Transcription'].map(lambda x: re.sub(r'[.|,|)|(|\|/]',r' ', str(x)))
msg['Text Transcription'] = msg['Text Transcription'].apply(lambda x: remove_stopwords(x))
msg.head(10)
```

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
sentence_model = SentenceTransformer('sentence-transformers/sentence-t5-base')
```

```python
texttransform=sentence_model.encode(msg['Text Transcription'])
```

```python
y=msg.iloc[:,0:5]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
X_train , X_test , y_train , y_test = train_test_split (texttransform, y , test_size = 0.20 , random_state = 0 )
```

```python
from keras.models import Sequential
from keras.layers import Dense
classifier=Sequential()
classifier.add(Dense(units = 20, kernel_initializer = 'uniform', activation = 'relu', input_dim = 768))
# Adding the second hidden layer
classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'softmax'))

# Compiling the ANN | means applying SGD on the whole ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train,y_train, batch_size = 10, epochs = 10)
```

```python
classifier.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = classifier.predict(X_test)
```
