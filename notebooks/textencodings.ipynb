```python
df=pd.read_csv("/content/drive/MyDrive/file/final/semevaltestnpy.csv")
```

```python
text=df["Text Transcription"].to_numpy()
```

```python
text.shape
```

```python
inputs = tokenizer(text[0], return_tensors="pt",add_special_tokens=True)
outputs = model(**inputs)
arrayclsxlnet=outputs[0][:,-1].detach().numpy()
arraymeanxlnet=outputs[0].mean(1).detach().numpy()
```

```python
for i in range(1,len(text)):
  inputs = tokenizer(text[i], return_tensors="pt",add_special_tokens=True)
  outputs = model(**inputs)
  arrayclsxlnet=np.append(arrayclsxlnet,outputs[0][:,-1].detach().numpy(),axis=0)
  arraymeanxlnet=np.append(arraymeanxlnet,outputs[0].mean(1).detach().numpy(),axis=0)
  print(i)
```

```python
arrayclsxlnet.shape,arraymeanxlnet.shape
```

```python
np.save("/content/drive/MyDrive/file/textencodings/xlnetclstest",arrayclsxlnet)
np.save("/content/drive/MyDrive/file/textencodings/xlnetmeantest",arraymeanxlnet)
```

```python
!pip install transformers
from transformers import AutoTokenizer, XLNetModel
import torch

tokenizer = AutoTokenizer.from_pretrained("xlnet-large-cased")
model = XLNetModel.from_pretrained("xlnet-large-cased")
```

```python
input_ids = torch.tensor(tokenizer.encode(text[0], add_special_tokens=True)).unsqueeze(0)  # Batch size 1

# Get the embeddings for the input text
with torch.no_grad():
    outputs = model(input_ids)
    arrayclslarge = outputs[0][:,-1].detach().numpy()
    arraymeanlarge=outputs[0].mean(1).detach().numpy()
```

```python
arrayclslarge.shape,arraymeanlarge.shape
```

```python
for i in range(1,len(text)):
  inputs = torch.tensor(tokenizer.encode(text[i], add_special_tokens=True)).unsqueeze(0)
  with torch.no_grad():
    outputs = model(inputs)
    arrayclslarge=np.append(arrayclslarge,outputs[0][:,-1].detach().numpy(),axis=0)
    arraymeanlarge=np.append(arraymeanlarge,outputs[0].mean(1).detach().numpy(),axis=0)
    print(i)
```

```python
np.save("/content/drive/MyDrive/file/textencodings/xlnetlargeclstest",arrayclslarge)
np.save("/content/drive/MyDrive/file/textencodings/xlnetlargemeantest",arraymeanlarge)
```
