```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/gdrive/MyDrive/file/final/semevalnpy.csv")
```

```python
df
```

```python
images=df["file_name"].to_numpy()
```

```python
!pip install transformers
from PIL import Image
from transformers import AutoProcessor, CLIPModel

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
folder="//gdrive/MyDrive/file/test/testinpainting/"
```

```python

image = Image.open(folder+"15001.jpg")

inputs = processor(images=image, return_tensors="pt")

image_features = model.get_image_features(**inputs).detach().numpy()
```

```python
image_features.shape
```

```python
for i in range(1,len(images)):
  image = Image.open(folder+images[i])
  inputs = processor(images=image, return_tensors="pt")
  features = model.get_image_features(**inputs).detach().numpy()
  image_features=np.append(image_features,features,axis=0)
  print(i)
```

```python
image_features.shape
```

```python
np.save("/gdrive/MyDrive/file/textencodings/clipmodeltestenc",image_features)
```

```python
text=df["Text Transcription"].to_numpy()
```

```python
text.shape
```

```python
from transformers import AutoTokenizer, CLIPModel

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
tokenizer = AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")

```

```python
from transformers import CLIPTextConfig, CLIPVisionConfig
from transformers import CLIPConfig, CLIPModel
# Initializing a CLIPText and CLIPVision configuration
config_text = CLIPTextConfig(max_position_embeddings=512)
config_vision = CLIPVisionConfig()

config = CLIPConfig.from_text_vision_configs(config_text, config_vision)
```

```python
model = CLIPModel(config)
```

```python
tokenizer = AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")
```

```python
inputs = tokenizer(text[0], padding=True, return_tensors="pt")
text_features = model.get_text_features(**inputs).detach().numpy()
```

```python
text_features.shape
```

```python
for i in range(1,len(text)):
  inputs = tokenizer(text[i], padding=True, return_tensors="pt")
  text_features=np.append(text_features,model.get_text_features(**inputs).detach().numpy(),axis=0)
  print(i)
```

```python
np.save("/gdrive/MyDrive/file/textencodings/cliptexttestenc",text_features)
```
